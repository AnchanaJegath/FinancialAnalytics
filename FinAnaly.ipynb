**The distributions of the Mar Cap - Crore in the data set and calculating their central tendencies. Visualising using a Histogram**

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Mar Cap - Crore', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Mar Cap (in crore)")
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Frequency')
plt.vlines(data['Mar Cap - Crore'].mean(), ymin = 0, ymax = 450, colors='pink', label='Mean')
plt.vlines(data['Mar Cap - Crore'].median(), ymin = 0, ymax = 450, colors='black', label='Median')
plt.legend()
plt.show()
```

```python
#Calculating the mean of Mar Cap feature
round(data['Mar Cap - Crore'].mean(),2)
```

```python
data['Mar Cap - Crore'].median()
```

```python
plt.figure(figsize = (9,4))

sns.distplot(data['Mar Cap - Crore'], color = 'blue')

plt.title("Distplot of Mar Cap - Crore ")
plt.xlabel('Market Captilization')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Mar Cap - Crore'].skew()
```

**The distributions of the Sales Qtr - Crore in the data set and calculating their central tendencies. Visualising using a Histogram**

```python
plt.figure(figsize = (9,4))

sns.histplot(data = data ,x = 'Sales Qtr - Crore', color = 'blue',
             edgecolor = 'linen', alpha = 0.7, bins = 5)

plt.title("Histogram of Sales Qtr(in crore)")
plt.xlabel('Sales Quatre (Crore)')
plt.ylabel('Frequency')
plt.vlines(data['Sales Qtr - Crore'].mean(), ymin = 0, ymax = 350, colors='pink', label='Mean')
plt.vlines(data['Sales Qtr - Crore'].median(), ymin = 0, ymax = 350, colors='black', label='Median')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (9,4))

sns.distplot(data['Sales Qtr - Crore'], color = 'blue')

plt.title("Distplot of Sales Qtr - Crore ")
plt.xlabel('Sales Qtr')
plt.ylabel('Frequency')
plt.show()
```

```python
data['Sales Qtr - Crore'].skew()
```

```python
sns.scatterplot(data=data, x='Mar Cap - Crore', y='Sales Qtr - Crore', palette='viridis',hue='Name',s=100, legend=False)

# Add labels and title
plt.title('Scatter Plot: Mar Cap vs. Sales Qtr')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Sales Qtr (Crore)')

# Show the plot
plt.show()
```
